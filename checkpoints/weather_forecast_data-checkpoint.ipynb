```python
from GQNN.data.dataset import Data_Read as dr
from GQNN.data.rfe import FeatureSelector
from GQNN.models.data_split import DataSplitter
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
df = dr.Read_csv('dataset/weather_forecast_data.csv',convert_strings=True)
sacled_df = dr.Scale_data()
```

```python
import seaborn as sns

sns.heatmap(df.corr(),annot=True)
```

```python
import matplotlib.pyplot as plt
plt.title("Checking for any Null values")
plt.plot(df.isnull())
plt.show()
```

# RFE 

```python
x = df.drop('Rain_rain',axis=1)
y = df['Rain_rain'].astype(int)
```

```python
x_train, x_test, y_train, y_test = DataSplitter(x,y,train_size=0.8,shuffle=False,random_state=42).split()
```

```python
rfe = FeatureSelector(
    estimator=LogisticRegression(solver='lbfgs',max_iter=10000),
    task='classification',
    step=5,
    cv=5
    
)

rfe.fit(x_train,y_train)

print(rfe.get_selected_features())
print(rfe.get_ranking())
```

```python
x = df.drop(['Cloud_Cover','Pressure','Rain_rain'],axis=1)
y = df['Rain_rain']
```

```python
x_train, x_test, y_train, y_test = DataSplitter(x,y,train_size=0.8,shuffle=False,random_state=42).split()
```

# Model Training

```python
model = LogisticRegression(solver='lbfgs',max_iter=10000)
model.fit(x_train,y_train)
```

# Model Evaluation

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
y_pred = model.predict(x_test)

acc = accuracy_score(y_test,y_pred)
clas = classification_report(y_test,y_pred)
print("Model Accuracy:",acc*100,"%")
print(f"Classification Report\n {clas}")
```

# Model Saving 

```python
from joblib import dump

model_name = "weather_forecast_data.joblib"
dump(model,model_name)
```
